# Kidney Disease DataSet

```python
X_data, y_data = preprocessing(kidney_path)

print(f"Number of features: {X_data[0].shape[0]}")
n_features = X_data[0].shape[0]
```

## Neural Networks
### We train an MLP model k times. Each time the dataset is randomly split for cross validation.

```python
k=10
out_dim = 2  # Binary classification 
hidden_dim = 10
epochs = 6 
lr = 0.001

accuracy_list = []
for i in range(k):
    X_train, y_train, X_test, y_test = dataset_loader(X_data, y_data, test_size=0.2)

    in_dim = X_train[0].shape[0]  # replace with your input dimension
    model = MLPModel(in_dim, out_dim, hidden_dim)
    model = train_model(model, X_train, y_train, learning_rate=lr, epochs=epochs, verbose=False)

    accuracy_list.append(model_performance(model, X_test, y_test))

accuracy_list = np.array(accuracy_list)
print(f"\nFinal results:")
print(f"Mean accuracy: {100*np.mean(accuracy_list):.2f}%")
print(f"Standard deviation: {100*np.std(accuracy_list):.2f}%")
```

## Decision Tree
### We train an Decision Tree choosing the optimal depth via cross validation and maximal accuracy.

```python
test_size = 0.3

min_depth = 2
max_depth = 15
depths = np.arange(min_depth, max_depth)

K = 200

depth_accuracy = []
for depth in depths:
    accuracy_list = []
    for k in range(K):
        X_train, y_train, X_test, y_test = dataset_loader(X_data, y_data, test_size=test_size,one_hot_encoding=False)
        dec_tree=Decision_tree(X_train, y_train, max_depth=depth)
        y_pred = dec_tree.predict(X_test)
        accuracy_list.append(accuracy_score(y_pred, y_test))
    depth_accuracy.append(100*np.mean(np.array(accuracy_list)))

plt.plot(depths, depth_accuracy)
plt.xlabel('Max depth of the tree', size=20)
plt.ylabel('Accuracy', size=20)


```
